LANGKAH 1

```python
# Step 1: Preprocessing Data

# 1. Kumpulkan dataset
import pandas as pd

# Load dataset
file_path = '/mnt/data/netflix_titles.csv'
df = pd.read_csv('netflix_titles.csv')

# Menampilkan 5 baris pertama untuk memahami struktur data
print("First 5 rows of the dataset:")
print(df.head())

# 2. Bersihkan data (hilangkan missing value dan outlier)

# Menampilkan informasi tentang missing values
print("\nChecking for missing values:")
print(df.isnull().sum())

# Menghilangkan missing value (Anda bisa memilih kolom yang dianggap penting)
df_clean = df.dropna()

print("\nData setelah menghapus missing values:")
print(df_clean.isnull().sum())

# 3. Normalisasi/transformasi data (jika diperlukan)
# Misalnya kita akan menormalisasi kolom 'release_year' menggunakan MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_clean['release_year'] = scaler.fit_transform(df_clean[['release_year']])

# 4. Pilih fitur-fitur relevan
# Misalnya kita akan memilih fitur 'release_year', 'duration', dan 'rating'
features = ['release_year', 'duration', 'rating']
df_features = df_clean[features]

print("\nSelected features:")
print(df_features.head())
```

LANGKAH 2

```python
# Step 2: Clustering

from sklearn.cluster import KMeans
import numpy as np

# 1. Pisahkan 'duration' berdasarkan 'min' dan 'Seasons'
# For movies, extract numeric values in minutes
df_clean['duration_minutes'] = df_clean['duration'].apply(lambda x: int(x.replace(' min', '')) if 'min' in x else np.nan)

# For shows, extract the number of seasons (convert to numeric)
df_clean['duration_seasons'] = df_clean['duration'].apply(lambda x: int(x.replace(' Seasons', '')) if 'Seasons' in x else np.nan)

# Fill NaN values in 'duration_minutes' with 0 for shows, and vice versa
df_clean['duration_minutes'].fillna(0, inplace=True)
df_clean['duration_seasons'].fillna(0, inplace=True)

# Convert 'rating' to categorical codes using .loc
df_clean.loc[:, 'rating'] = pd.Categorical(df_clean['rating']).codes

# 2. Pilih fitur yang akan digunakan untuk clustering, termasuk 'release_year', 'duration_minutes', 'duration_seasons', dan 'rating'
df_features = df_clean[['release_year', 'duration_minutes', 'duration_seasons', 'rating']]

# 3. Tentukan jumlah klaster (K), misalnya K=3
k = 3
kmeans = KMeans(n_clusters=k)

# 4. Jalankan algoritma clustering
kmeans.fit(df_features)

# Simpan hasil klaster
df_clean['cluster'] = kmeans.labels_

print("\nCluster results:")
print(df_clean[['release_year', 'duration_minutes', 'duration_seasons', 'rating', 'cluster']].head())
```

LANGKAH 3

```python
# Step 3: Penggabungan dengan Machine Learning

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# 1. Pilih model machine learning (misalnya, Klasifikasi dengan Random Forest)
# Pastikan kita menggunakan kolom numerik ('release_year', 'duration_minutes', 'duration_seasons', 'rating', 'cluster')
X = df_clean[['release_year', 'duration_minutes', 'duration_seasons', 'rating', 'cluster']]
y = df_clean['cluster']  # Target adalah hasil klaster untuk contoh ini

# 2. Split data untuk pelatihan dan pengujian
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Latih model machine learning
model = RandomForestClassifier()
model.fit(X_train, y_train)

# 4. Evaluasi kinerja model
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print("\nModel accuracy with clustering as feature: {:.2f}%".format(accuracy * 100))
```

LANGKAH 4

```python
# Step 4: Evaluasi dan Pengoptimalan

from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV

# 1. Evaluasi kinerja model dengan clustering
print("\nClassification report with clustering:")
print(classification_report(y_test, y_pred))

# 2. Pengoptimalan hyperparameter
param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [5, 10]
}
grid_search = GridSearchCV(RandomForestClassifier(), param_grid, cv=5)
grid_search.fit(X_train, y_train)

print("\nBest hyperparameters:", grid_search.best_params_)

# 3. Latih model tanpa clustering (gunakan hanya fitur asli tanpa 'cluster')
X_nc = df_clean[['release_year', 'duration_minutes', 'duration_seasons', 'rating']]
X_train_nc, X_test_nc, y_train_nc, y_test_nc = train_test_split(X_nc, y, test_size=0.3, random_state=42)

model_nc = RandomForestClassifier()
model_nc.fit(X_train_nc, y_train_nc)

# 4. Evaluasi kinerja model tanpa clustering
y_pred_nc = model_nc.predict(X_test_nc)

print("\nClassification report without clustering:")
print(classification_report(y_test_nc, y_pred_nc))
```
